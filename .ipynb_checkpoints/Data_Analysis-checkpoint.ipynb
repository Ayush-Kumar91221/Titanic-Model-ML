```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./Data/train.csv")
data.head()
```

```python
data.info()
```

```python
def DataRefine(dataframe):
    dataframe.drop(["Cabin","Name"],axis=1,inplace=True)
    dataframe["Fare_Group"] = (dataframe["Fare"]/10).apply(np.ceil)
```

```python
DataRefine(data)
```

```python
data
```

```python
def StackedBarGraph(x,y,dataframe ):
    attributes = [x,y]
    data = dataframe[attributes].copy()
    data = data.groupby(attributes).size().unstack()
    sum = data.sum(axis=1)
    percentage = data.div(sum,axis=0)*100
    percentage.plot(kind="bar", stacked=True)
    plt.show()
```

```python
StackedBarGraph( dataframe = data,x= "Pclass",y="Survived")
```

```python
StackedBarGraph(dataframe = data, x= "Fare_Group",y = "Survived")
```

```python
dataset = data[["Survived","Age","Sex"]]
plt.figure(figsize=(10, 10)) 
sns.stripplot(dataset, x = "Sex", y="Age",hue="Survived",dodge=True)
```

```python
import numpy as np
import pandas as pd
import os
import math
import matplotlib.pyplot as plt
```

```python
# def Import_data():
#     Data_path = "" # to be specified
#     Data_location = "" # to be specified
#     if not os.path.isdir("Data"):
#         os.makedirs("Data")
train_set = pd.read_csv("./Data/train.csv")
train_set.head()
    
```

```python
train_set.info()
```

```python
train_set.drop("Cabin",axis=1,inplace=True)
train_set.drop("Name",axis=1,inplace=True)
```

```python
train_set.head()
```

```python
data= train_set.groupby(["Pclass","Survived"]).size().unstack()
print(data.shape)
data
```

```python
sum = data.sum(axis =1)
sum
```

```python
percentage =data.div(sum,axis=0)*100
percentage
```

```python
percentage.plot(kind="bar",stacked=True)
plt.xlabel = "Pclass"
plt.ylabel = "Percent"
```

```python
data = train_set[["Survived","Fare"]].copy()
data["Fare_group"]= (data["Fare"]/10).apply(np.ceil)
data= data.drop("Fare",axis=1).groupby(["Survived","Fare_group"]).size().unstack()
sum = data.sum(axis=0)
# sum
percentage = data.div(sum,axis=1)*100
percentage = percentage.transpose()
percentage.plot(kind="bar",stacked=True)
plt.show()
```

```python
data= train_set.groupby(["Sex","Survived"]).size().unstack()
sum = data.sum(axis=1)
percentage = data.div(sum,axis=0)*100
percentage.plot(kind="bar",stacked=True)
```
